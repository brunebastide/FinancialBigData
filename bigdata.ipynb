```python
import numpy as np
import pandas as pd
```

```python
path = 'data.csv'
```

```python
csv_file_data = pd.read_csv(path)
```

```python
csv_file_data.head()
```

```python
data_df = csv_file_data.drop('PERMNO', axis = 1).groupby(['date', 'COMNAM']).sum().unstack()
data_df
```

```python
data_df.shape
```

```python
data_df_clean = data_df.dropna(axis=1,how='all')
```

```python
data_df_clean.shape
```

```python
data_df_clean.to_csv('new_sample.csv', sep=',')
```
